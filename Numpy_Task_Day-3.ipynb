# Car Sales Prediction & Analytics System



## 1️ Introduction to NumPy

```python
import numpy as np
np.random.seed(42)
print('NumPy Version:', np.__version__)
```

## 2️ Creating Car Sales Dataset

```python
num_records = 1000
car_ids = np.arange(1, num_records + 1)
price = np.random.randint(500000, 2500000, num_records)
engine_size = np.random.randint(1000, 3000, num_records)
car_age = np.random.randint(0, 11, num_records)
units_sold = np.random.randint(1, 51, num_records)
dataset = np.column_stack((car_ids, price, engine_size, car_age, units_sold))
print('Dataset Shape:', dataset.shape)
dataset[:5]
```

## 3️ Array Properties

```python
print('Shape:', dataset.shape)
print('Dimensions:', dataset.ndim)
print('Size:', dataset.size)
print('Data Type:', dataset.dtype)
```

## 4️ Revenue Calculation (Mathematical Operations)

```python
revenue = price * units_sold
print('Sample Revenue:', revenue[:5])
```

## 5️ Statistical Functions

```python
print('Average Car Price:', np.mean(price))
print('Maximum Revenue:', np.max(revenue))
print('Minimum Revenue:', np.min(revenue))
print('Total Revenue:', np.sum(revenue))
print('Standard Deviation of Revenue:', np.std(revenue))
```

## 6️ Boolean Indexing

```python
high_sales = car_ids[units_sold > 40]
print('High Sales Records Count:', high_sales.size)
```

## 7️ Reshaping & Axis Operations

```python
reshaped_revenue = revenue[:100].reshape(10,10)
print('Row-wise Sum:', np.sum(reshaped_revenue, axis=1)[:3])
print('Column-wise Mean:', np.mean(reshaped_revenue, axis=0))
```

## 8️ Sorting & Top Performing Cars

```python
top5_revenue = np.sort(revenue)[-5:]
print('Top 5 Revenue Values:', top5_revenue)
```

## 9️ Broadcasting (Price Increase Simulation)

```python
increased_price = price + (price * 0.10)
print('Sample Increased Prices:', increased_price[:5])
```

## 10 Linear Algebra Example

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print('Matrix Multiplication:\n', np.dot(A,B))
print('Determinant of A:', np.linalg.det(A))
```

## 11 Saving & Loading Data

```python
np.save('car_sales_revenue.npy', revenue)
loaded_revenue = np.load('car_sales_revenue.npy')
print('Loaded Revenue Sample:', loaded_revenue[:5])
```

## 11 Basic Sales Prediction Insight

```python
predicted_units = units_sold * 1.05
predicted_revenue = price * predicted_units
print('Predicted Total Revenue Next Year: ₹', np.sum(predicted_revenue))
```
